**Game of Thrones** is an American fantasy drama television series created by David Benioff and D. B. Weiss for _HBO_. It is an adaptation of A Song of Ice and Fire, a series of fantasy novels by George R. R. Martin, the first of which is A Game of Thrones. The show premiered on _HBO_ in the United States on April 17, 2011, and concluded on May 19, 2019, with 73 episodes broadcast over eight seasons.

Set on the fictional continents of *Westeros* and *Essos*, **Game of Thrones** has a large ensemble cast and follows several story arcs throughout the course of the show. The first major arc concerns the Iron Throne of the Seven Kingdoms of Westeros through a web of political conflicts among the noble families either vying to claim the throne or fighting for independence from whoever sits on it. The second major arc focuses on the last descendant of the realm's deposed ruling dynasty, who has been exiled to Essos and is plotting to return and reclaim the throne. The third follows the Night's Watch, a military order defending the realm against threats from beyond Westeros' northern border.

```python
# Load the dataset
df = pd.read_csv(r"C:\Users\Alovsat Abdullayev\Desktop\UW\Lessons\Extra year\Research\Assignment\Game_of_Thrones.csv")

# Display the first few rows of the dataset
df.head()
```

```python
# Assuming df is your DataFrame
# Plotting IMDb Ratings
plt.figure(figsize=(10, 5))
plt.plot(df['No. of Episode (Overall)'], df['IMDb Rating'], marker='o')
plt.title('IMDb Ratings Over Episodes')
plt.xlabel('Episode')
plt.ylabel('IMDb Rating')
plt.grid(True)
plt.show()
```

The ratings for **"Game of Thrones"** episodes were generally quite high, with many episodes receiving ratings around 9 or above, indicating strong positive reception. There's a noticeable consistency in high ratings across the span of episodes, showcasing the show's ability to maintain quality and viewer engagement, according to IMDb users.
However, there is a significant dip towards the end of the series, where the ratings drop sharply. This is reflective of the general viewer sentiment regarding the final season, where many fans expressed disappointment with the storytelling and character arcs' resolutions.

```python
# Bar chart for Viewers
plt.figure(figsize=(10, 5))
plt.bar(df['No. of Episode (Overall)'], df['U.S. Viewers (Millions)'])
plt.title('U.S. Viewers (Millions) Per Episode')
plt.xlabel('Episode')
plt.ylabel('U.S. Viewers (Millions)')
plt.show()
```

There's a notable increase especially towards the later episodes, with the final episodes reaching the peak viewership, each surpassing 12 million U.S. viewers. This pattern reflects strong viewer engagement and anticipation for the climax of the series. The growing numbers could also suggest that more viewers were catching up with the series, leading to higher numbers as it approached its conclusion.

```python
# Multiple line plots for Rotten Tomatoes and Metacritic ratings
plt.figure(figsize=(10, 5))
plt.plot(df['No. of Episode (Overall)'], df['Rotten Tomatoes Rating (Percentage)'], label='Rotten Tomatoes Rating (%)', marker='*')
plt.plot(df['No. of Episode (Overall)'], df['Metacritic Ratings'], label='Metacritic Ratings', marker='x')
plt.title('Comparison of Rotten Tomatoes and Metacritic Ratings')
plt.xlabel('Episode')
plt.ylabel('Ratings')
plt.legend()
plt.grid(True)
plt.show()
```

**Rotten Tomatoes Ratings**: The Rotten Tomatoes ratings (depicted in blue) are consistently high throughout the series, mostly staying above 80%. This suggests a generally positive reception from critics aggregated by Rotten Tomatoes. The trendline for Rotten Tomatoes ratings displays a few dips, but these are relatively minor until the final episodes, where there is a significant drop, reflecting a less favorable critical response.

**Metacritic Ratings**: The Metacritic ratings (depicted in orange), represented by an average score given by critics, show a very consistent line at the lower end of the graph, staying around the same mark throughout the series. This plot points to a more uniform critical response on Metacritic, without the sharp decline seen in the Rotten Tomatoes data.
